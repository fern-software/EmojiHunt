```python
import numpy as np
import matplotlib.pyplot as plt

# We want bins of 5 for our histogram
bins = np.arange(0, max_score, 5)

# Plot
_, ax = plt.subplots()
ax.hist(scores, bins=bins)
ax.set_title('Scores for official test run')
ax.set_xlabel('Score (binned)')
ax.set_ylabel('Frequency')

# Add an annotation for the final score
ax.annotate(f'Final score: {final_score:.2f}', xy=(final_score, 0), xytext=(final_score, 5), arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
